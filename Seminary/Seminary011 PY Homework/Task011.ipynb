###### f(x) = 18x^3+5x^2 + 10x - 30
***
1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0
***
Необязательно, для интереса:
f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30

```python
from sympy import *
```

```python
#1. Определить корни
x = Symbol('x')
f = 18*(x**3) + 5*(x**2) + 10*x -30
ans = solve(f,x) 
print(ans)
print()

for res in ans:
    print(res.evalf())
```

```python
#2. Найти интервалы, на которых функция возрастает
#3. Найти интервалы, на которых функция убывает
print(solve(f > 0, x))
print(solve(f < 0, x))
```

```python
#4. Построить график
plot(f)
plot(f,(x,-1.2,1.2))
```

```python
#5. Вычислить вершину
df = diff(f,x)
print(solve(df,x))
```

```python
#6. Определить промежутки, на котором f > 0
#7. Определить промежутки, на котором f < 0
print(solve(df>0,x))
print(solve(df<0,x))
```
